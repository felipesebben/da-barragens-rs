```python
import requests
import pandas as pd
```

```python
url = "https://servicodados.ibge.gov.br/api/v3/agregados/4714/periodos/-6/variaveis/93?localidades=N6[N3[43]]"

data = requests.get(url).json()
df = pd.DataFrame(data)
```

```python
df
```

```python
df_final = pd.DataFrame()

# Iterate through each item in the data list
for item in data:
    # Extract variable and unit
    variavel = item["variavel"]
    unidade = item["unidade"]

    # Extract the series data
    for entry in item["resultados"][0]["series"]:
        # Flatten the localidade information
        localidade_id = entry["localidade"]["id"]
        localidade_nome = entry["localidade"]["nome"]

        # Iterate through each year in the serie
        for year, value in entry["serie"].items():
            # Create a DataFrame from the extracted data
            df_temp = pd.DataFrame(
                {
                    "id": [localidade_id],
                    "nome": [localidade_nome],
                    "year": [year],
                    "value": [value],
                    "variavel": [variavel],
                    "unidade": [unidade],
                }
            )
            # Append to the final DataFrame
            df_final = pd.concat([df_final, df_temp], ignore_index=True)

# Print the final DataFrame
print(df_final)
```
